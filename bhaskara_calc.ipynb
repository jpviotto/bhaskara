```python
import pandas as pd
import numpy as np
import sympy as sp 
from sympy.plotting import *
```

```python
x, y = symbols('x y')
```

```python
class bhaskara:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b    
        self.c = c
    
    
    def delta(self):
        return (self.b ** 2) - (4 * self.a * self.c)

  
    def x1(self):
        return (-self.b + np.sqrt(self.b ** 2 - (4 * self.a * self.c))) / (2 * self.a)
   

    def x2(self):
        return (-self.b - np.sqrt(self.b ** 2 - (4 * self.a * self.c))) / (2 * self.a)
    
    def x_vertice(self):
        return (-self.b) / (2 * self.a)
    
    def y_vertice(self):
        return ((self.b ** 2) - (4 * self.a * self.c)) / (4 * self.a)
        
```

```python
resposta = True

while resposta:
    
    'Bhaskara'


    a = float(input('Informe o coeficiente a: '))
    b = float(input('Informe o coeficiente b: '))
    c = float(input('Informe o coeficiente c: '))
    
    print('\n')
    
    equacao1 = bhaskara(a, b, c)
    
    delta =  equacao1.delta()
    
    if delta < 0:
        print('delta: {:.2f}'.format(delta))
        print('Não há raízes reais')
    elif delta > 0:
        x1 = equacao1.x1()
        x2 = equacao1.x2()
        x_v = equacao1.x_vertice()
        y_v = equacao1.y_vertice()
        print('delta: {:.2f}'.format(delta))
        print('x1: {:.2f}'.format(x1))
        print('x2: {:.2f}'.format(x2))
        print('x_vértice: {:.2f}'.format(x_v))
        print('y_vértice: {:.2f}'.format(y_v))
    else:
        print('delta: {:.2f}'.format(delta))
        x1 = equacao1.x1()
        x2 = equacao1.x2()
        x_v = equacao1.x_vertice()
        y_v = equacao1.y_vertice()
        print('x1: {:.2f} = x2:{:.2f}'.format(x1, x2))
        print('x_vértice: {:.2f}'.format(x_v))
        print('y_vértice: {:.2f}'.format(y_v))
    
    
    print('\n')
    
    
    plot(a * x**2 + b * x + c)
    
    
    print('\n')
    
          
    continuar = str(input('Deseja calcular outra função (s/n): ')).upper()
    
    
    if continuar == 'S':
        resposta = True
    else:
        resposta = False
      
   
```
